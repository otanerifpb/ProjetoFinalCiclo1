# Classificação

1) Obter um dataset de classificação e outro de regressão, diferentes dos
já apresentados durante a capacitação. Os datasets podem ser obtidos
dos seguintes locais:

```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"

# Carregar os dados do dataset
df = pd.read_csv(url, header=None)

# Verificar a quantidade de colunas do banco, para usar nas Features(X)
columns = len(df.columns)

df
```

2) Analisar o dataset usando os métodos do Pandas. Comparar as
features e verificar se há valores nulos, correspondência entre features,
converter os labels para números e features categóricas usar One-Hot
Encoder. Normalizar as features na escala 0..1.

```python
# Para contar a quantidade de valores iguais de uma coluna 'NaN'
#coluna = df[0] # Contem 0 Nan
#coluna = df[1] # Contem 0 Nan
#coluna = df[2] # Contem 0 Nan
#coluna = df[3] # Contem 0 Nan
#coluna = df[4] # Contem 0 Nan
#coluna = df[5] # Contem 0 Nan
#coluna = df[6] # Contem 0 Nan
#coluna = df[7] # Contem 0 Nan
#coluna = df[8] # Contem 0 Nan
#coluna = df[9] # Contem 0 Nan
#coluna = df[10] # Contem 0 Nan
#coluna = df[11] # Contem 0 Nan
#coluna = df[12] # Contem 0 Nan
coluna = df[13] # Contem 0 Nan
cont_nan = coluna.isna().sum()

print("Total NaN: " + str(cont_nan))
```

```python
from sklearn.preprocessing import MinMaxScaler

# A colula Target(y) é a coluna com 'id'
y = df[0]

# Separar as demais colunas para Features(X), a partir da coluna 1
X = df.loc[:,1:columns-1]

# Normalizando todo o 'X' de uma só vez
X[df.columns[1:]] = MinMaxScaler().fit_transform(df[df.columns[1:]])

X
```

4) Para o dataset de classificação:\
4.1) Separar os dados de treinamento e teste (80% e 20%);

```python
from sklearn.model_selection import train_test_split

# Realizar separação Treinamento/Teste com Classificação
res = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=None, stratify = y)
X_train, X_test, y_train, y_test = res
```

4.2) Treinar os modelos:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

# Árvore de Decisão para Classificação
modelC1 = DecisionTreeClassifier(criterion = "entropy")
modelC2 = DecisionTreeClassifier(criterion = "entropy", max_depth = 3)

# K-NN para Classificação
modelC3 = KNeighborsClassifier(n_neighbors = 7, metric = 'euclidean', algorithm = 'brute')
modelC4 = KNeighborsClassifier(n_neighbors = 5, metric = 'manhattan', algorithm = 'brute')

# Redes Neurais para Classificação
modelC5 = MLPClassifier(hidden_layer_sizes=(2,2), activation='tanh', max_iter=5000)
modelC6 = MLPClassifier(hidden_layer_sizes=(4,5), activation='relu', max_iter=2000)

# SVM para Classificação
modelC7 = SVC()
modelC8 = SVC(gamma="auto", kernel="rbf")

# Random Forest para Classificação
modelC9 = RandomForestClassifier(n_estimators=100)

# GradientBoost para Classificação
modelC10 = GradientBoostingClassifier(n_estimators=100)
```

4.3) Executar 10 vezes o Train-Test-Split com treinamento e
predição dos modelos para cada conjunto de dados randomicamente
selecionados;

```python
from sklearn import metrics

# Treinameto para Classificação
treinoC1 = modelC1.fit(X_train, y_train)
treinoC2 = modelC2.fit(X_train, y_train)
treinoC3 = modelC3.fit(X_train, y_train)
treinoC4 = modelC4.fit(X_train, y_train)
treinoC5 = modelC5.fit(X_train, y_train)
treinoC6 = modelC6.fit(X_train, y_train)
treinoC7 = modelC7.fit(X_train, y_train)
treinoC8 = modelC8.fit(X_train, y_train)
treinoC9 = modelC9.fit(X_train, y_train)
treinoC10 = modelC10.fit(X_train, y_train)

# Resultados dos Testes
predictionsC1 = treinoC1.predict(X_test)
predictionsC2 = treinoC2.predict(X_test)
predictionsC3 = treinoC3.predict(X_test)
predictionsC4 = treinoC4.predict(X_test)
predictionsC5 = treinoC5.predict(X_test)
predictionsC6 = treinoC6.predict(X_test)
predictionsC7 = treinoC7.predict(X_test)
predictionsC8 = treinoC8.predict(X_test)
predictionsC9 = treinoC9.predict(X_test)
predictionsC10 = treinoC10.predict(X_test)

# Para mostrar a acurácia dos Testes em '%
accC1 = metrics.accuracy_score(predictionsC1, y_test)
accC2 = metrics.accuracy_score(predictionsC2, y_test)
accC3 = metrics.accuracy_score(predictionsC3, y_test)
accC4 = metrics.accuracy_score(predictionsC4, y_test)
accC5 = metrics.accuracy_score(predictionsC5, y_test)
accC6 = metrics.accuracy_score(predictionsC6, y_test)
accC7 = metrics.accuracy_score(predictionsC7, y_test)
accC8 = metrics.accuracy_score(predictionsC8, y_test)
accC9 = metrics.accuracy_score(predictionsC9, y_test)
accC10 = metrics.accuracy_score(predictionsC10, y_test)

# Formatação com 3 casas decimais para Acurácia
showC1 = "{:.3f}".format(accC1*100)
showC2 = "{:.3f}".format(accC2*100)
showC3 = "{:.3f}".format(accC3*100)
showC4 = "{:.3f}".format(accC4*100)
showC5 = "{:.3f}".format(accC5*100)
showC6 = "{:.3f}".format(accC6*100)
showC7 = "{:.3f}".format(accC7*100)
showC8 = "{:.3f}".format(accC8*100)
showC9 = "{:.3f}".format(accC9*100)
showC10 = "{:.3f}".format(accC10*100)

print('Modelo Classificação Árvore Decisão')
print("Árvore Decisão Profundidade livre " + "Acurácia: {}%".format(showC1))
print("Árvore Decisão Profundidade = 3 " + "Acurácia: {}%".format(showC2))
print('*'*60)
print('Modelo Classificação K-NN')
print("Classificação 7-NN + Euclidean  " + "Acurácia: {}%".format(showC3))
print("Classificação 5-NN + Mamhattan " + "Acurácia: {}%".format(showC4))
print('*'*60)
print('Modelo Classificação MLP')
print("Classificação MLP 2x2 + tanh  " + "Acurácia: {}%".format(showC5))
print("Classificação MLP 4x5 + relu " + "Acurácia: {}%".format(showC6))
print('*'*60)
print('Modelo Classificação SVM')
print("Classificação SVM  " + "Acurácia: {}%".format(showC7))
print("Classificação SVM + kernel " + "Acurácia: {}%".format(showC8))
print('*'*60)
print('Modelo Classificação Random Forest')
print("Classificação Random Forest  " + "Acurácia: {}%".format(showC9))
print('*'*60)
print('Modelo Classificação Gradient Boost')
print("Classificação Gradient Boost  " + "Acurácia: {}%".format(showC10))
```

1) Obter um dataset de classificação e outro de regressão, diferentes dos

dos seguintes locais:

```python
import seaborn as sns
from sklearn import datasets, linear_model

# Carregar os dados do dataset
#df1 = sns.load_dataset("mpg")
df1 = sns.load_dataset("healthexp")

# Verificar a quantidade de colunas do banco, para usar nas Features(X)
columns = len(df1.columns)

df1
```

2) Analisar o dataset usando os métodos do Pandas. Comparar as
features e verificar se há valores nulos, correspondência entre features,
converter os labels para números e features categóricas usar One-Hot
Encoder. Normalizar as features na escala 0..1.

```python
# Para contar a quantidade de valores iguais de uma coluna 'NaN'
#coluna = df1["Year"] # Contem 0 NaN
#coluna = df1["Country"] # Contem 0 NaN
#coluna = df1["Spending_USD"] # Contem 0 NaN
coluna = df1["Life_Expectancy"] # Contem 0 NaN

cont_nan = coluna.isna().sum()

print("Total NaN: " + str(cont_nan))
df1.info()
```

```python
# 'dropna' remove as linhas que contem NaN como valores
df1 = df1.dropna()

df1.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
import pandas as pd

df1 = pd.DataFrame(data = df1)
# One Hot Encoder usado para as Features(X), onde cada valor da coluna vai gerar uma nova coluna, onde esta fica com código binário
one_hot_encoder = make_column_transformer((OneHotEncoder(), ['Country']), remainder='passthrough')
#one_hot_encoder = pd.get_dummies(df, columns = ['subject','event', 'region'])

# Transforando os valores categóricos para one_hot_encoder em todo df
df1 = one_hot_encoder.fit_transform(df1)

# Para visualizar o banco de uma forma estilizada do Pandas
df1 = pd.DataFrame(data = df1)

df1
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Normalizando todo o 'X' de uma só vez quando a Target(y) é a coluna '0'
df1[6] = MinMaxScaler().fit_transform(np.array(df1[6]).reshape(-1,1))
df1[7] = MinMaxScaler().fit_transform(np.array(df1[7]).reshape(-1,1))

df1
```

```python
# Separar a coluna Target(y), usando o índice da coluna
y1 = df1[8].copy()

# Separar as demais colunas para Features(X)
#X1 = df1.loc[:,0:8]

# Separar as demais colunas para Features(X), exceto uma determinada coluna
X1 = df1.drop(8, axis=1).copy()

print(y1.head(10))
print()
X1.head(10)
```

5) Para o dataset de regressão:\
5.1) Separar os dados de treinamento e teste (80% e 20%);

```python
from sklearn.model_selection import train_test_split

# Realizar separação Treinamento/Teste com Classificação
res1 = train_test_split(X1, y1, train_size=0.8, test_size=0.2, random_state=None)
X1_train, X1_test, y1_train, y1_test = res1
```

5.2) Treinar os modelos:

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

# Model Árvore Decisão para Regressão
modelR1 = DecisionTreeRegressor()
modelR2 = DecisionTreeRegressor(max_depth = 5)

# Model K-NN para Regressão
modelR3 = KNeighborsRegressor(n_neighbors = 3)
modelR4 = KNeighborsRegressor(n_neighbors = 5)

# Model MLP para Regressão
modelR5 = MLPRegressor(hidden_layer_sizes=(3,3), activation='relu', max_iter=40000)
modelR6 = MLPRegressor(hidden_layer_sizes=(3,5), activation='relu', max_iter=30000)

# Model SVM para Regressão
param_grid7 = {'C': [1], 'epsilon': [0.01]}
modelR7 = GridSearchCV(SVR(), param_grid7, cv=5)
param_grid8 = {'C': [10], 'epsilon': [0.001]}
modelR8 = GridSearchCV(SVR(), param_grid8, cv=5)

# Model Random Forest para Regressão
modelR9 = RandomForestRegressor(n_estimators=100, max_depth=5)

# Model Gradient Boost para Regressão
modelR10 = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=10)
```

5.3) Executar 10 vezes o Train-Test-Split com treinamento e
predição dos modelos para cada conjunto de dados randomicamente
selecionados;

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Treinameto para Regressão
treinoR1 = modelR1.fit(X1_train, y1_train)
treinoR2 = modelR2.fit(X1_train, y1_train)
treinoR3 = modelR3.fit(X1_train, y1_train)
treinoR4 = modelR4.fit(X1_train, y1_train)
treinoR5 = modelR5.fit(X1_train, y1_train)
treinoR6 = modelR6.fit(X1_train, y1_train)
treinoR7 = modelR7.fit(X1_train, y1_train)
treinoR8 = modelR8.fit(X1_train, y1_train)
treinoR9 = modelR9.fit(X1_train, y1_train)
treinoR10 = modelR10.fit(X1_train, y1_train)

# Resultados dos Testes
predictionsR1 = treinoR1.predict(X1_test)
predictionsR2 = treinoR2.predict(X1_test)
predictionsR3 = treinoR3.predict(X1_test)
predictionsR4 = treinoR4.predict(X1_test)
predictionsR5 = treinoR5.predict(X1_test)
predictionsR6 = treinoR6.predict(X1_test)
predictionsR7 = treinoR7.predict(X1_test)
predictionsR8 = treinoR8.predict(X1_test)
predictionsR9 = treinoR9.predict(X1_test)
predictionsR10 = treinoR10.predict(X1_test)

# Formatação com 3 casas decimais para Erro
erro1 = "{:.3f}".format(mean_squared_error(y1_test, predictionsR1))
erro2 = "{:.3f}".format(mean_squared_error(y1_test, predictionsR2))
erro3 = "{:.3f}".format(mean_squared_error(y1_test, predictionsR3))
erro4 = "{:.3f}".format(mean_squared_error(y1_test, predictionsR4))
erro5 = "{:.3f}".format(mean_squared_error(y1_test, predictionsR5))
erro6 = "{:.3f}".format(mean_squared_error(y1_test, predictionsR6))
erro7 = "{:.3f}".format(mean_squared_error(y1_test, predictionsR7))
erro8 = "{:.3f}".format(mean_squared_error(y1_test, predictionsR8))
erro9 = "{:.3f}".format(mean_squared_error(y1_test, predictionsR9))
erro10 = "{:.3f}".format(mean_squared_error(y1_test, predictionsR10))

# Para mostrar a acurácia dos Testes em '%
accR1 = modelR1.score(X1_test, y1_test)
accR2 = modelR2.score(X1_test, y1_test)
accR3 = modelR3.score(X1_test, y1_test)
accR4 = modelR4.score(X1_test, y1_test)
accR5 = modelR5.score(X1_test, y1_test)
accR6 = modelR6.score(X1_test, y1_test)
accR7 = modelR7.score(X1_test, y1_test)
accR8 = modelR8.score(X1_test, y1_test)
accR9 = modelR9.score(X1_test, y1_test)
accR10 = modelR10.score(X1_test, y1_test)

# Formatação com 3 casas decimais para Acurácia
showR1 = "{:.3f}".format(accR1*100)
showR2 = "{:.3f}".format(accR2*100)
showR3 = "{:.3f}".format(accR3*100)
showR4 = "{:.3f}".format(accR4*100)
showR5 = "{:.3f}".format(accR5*100)
showR6 = "{:.3f}".format(accR6*100)
showR7 = "{:.3f}".format(accR7*100)
showR8 = "{:.3f}".format(accR8*100)
showR9 = "{:.3f}".format(accR9*100)
showR10 = "{:.3f}".format(accR10*100)

print('Modelo Regreção Árvore Decisão')
print('Resultado Erro Profundidade livre: ' + str(erro1))
print('Resultado Acurácia Profundidade livre: ' + '{}%'.format(showR1))
print('Resultado Erro Profundidade = 5: ' + str(erro2))
print('Resultado Acurácia Profundidade = 5: ' + '{}%'.format(showR2))
print('*'*60)
print('Modelo Regreção K-NN')
print('Resultado Erro 3-NN: ' + str(erro3))
print('Resultado Acurácia 3-NN: ' + '{}%'.format(showR3))
print('Resultado Erro 5-NN: ' + str(erro4))
print('Resultado Acurácia 5-NN: ' + '{}%'.format(showR4))
print('*'*60)
print('Modelo Regreção MLP')
print('Resultado Erro MLP 3x2: ' + str(erro5))
print('Resultado Acurácia MLP 3x2: ' + '{}%'.format(showR5))
print('Resultado Erro MLP 3x5 ' + str(erro6))
print('Resultado Acurácia MLP 3x5: ' + '{}%'.format(showR6))
print('*'*60)
print('Modelo Regreção SVM')
print('Resultado Erro SVM C=1: ' + str(erro7))
print('Resultado Acurácia SVM C=1: ' + '{}%'.format(showR7))
print('Resultado Erro SVM C=10 ' + str(erro8))
print('Resultado Acurácia SVM C=10: ' + '{}%'.format(showR8))
print('*'*60)
print('Modelo Regreção Random Forest')
print('Resultado Erro Random Forest: ' + str(erro9))
print('Resultado Acurácia Random Forest: ' + '{}%'.format(showR9))
print('*'*60)
print('Modelo Regreção Gradient Boost')
print('Resultado Erro Gradient Boost: ' + str(erro10))
print('Resultado Acurácia Gradient Boost: ' + '{}%'.format(showR10))
```
